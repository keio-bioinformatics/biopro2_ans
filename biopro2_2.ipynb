## 課題1

文字列"animal"の文字を逆に（末尾から先頭に向かって）並べた文字列を得よ．

>***ヒント*** 
- スライスを用いて要素を取得する際に、ステップ数を同時に指定する事ができる．逆向きにする場合は負の数を使用する．
    ```python
    >>> a = [1,2,3,4,5,6,7,8,9]
    >>> a[1:8:2]          # インデックス1〜7の要素を2ステップで
    [2, 4, 6, 8]
    ```
- pythonはシンプルな記法に立脚した、高い可読性(保守性)を実現できる言語である．より簡潔なコーディングを模索してみよう．  
"There should be one– and preferably only one –obvious way to do it."


````python
animal = "animal"
reversed_animal = animal[::-1]
print(reversed_animal)
````

## 課題2

「パトカー」＋「タクシー」の文字を先頭から交互に連結して文字列「パタトクカシーー」を得よ．


````python
pat = "パトカー"
taxi = "タクシー"
combined = ""

# zipを使用して、それぞれの文字を交互に連結
for p, t in zip(pat, taxi):
    combined += p + t

print(combined)
````

## 課題3
ネストされたリストが与えられている。各サブリストには数値のリストと文字列を含むリストが混在している。サブリストの中で、すべての要素が数値であるリストだけを抽出し、それらのサブリストの合計を計算し、合計値が10以上のサブリストのみを新しいリストに格納せよ。

```python
data = [[1, 2, 3], [4, 5], ['a', 'b', 'c'], [6, 7, 8, 9], [2, 'three', 5], [5, 5]]
```

期待される出力:
```
[[6, 7, 8, 9], [5, 5]]
```

### ヒント
- `isinstance()`関数を使用して、要素が数値かどうかを確認できる。
- 例えば、次のコードは、`x`が数値（整数または浮動小数点数）かどうかを判定する:
  ```python
  x = 5
  if isinstance(x, (int, float)):
      print("xは数値である")
  else:
      print("xは数値ではない")
  ```
- 上記の例では、`x`が数値の場合に「xは数値である」と表示される。
- サブリスト内の要素を1つずつ確認し、数値以外の要素が含まれている場合はそのサブリストを無視すること。

````python
data = [[1, 2, 3], [4, 5], ['a', 'b', 'c'], [6, 7, 8, 9], [2, 'three', 5], [5, 5]]

result = []

# すべてのサブリストを順に処理
for lst in data:
    is_numeric = True
    total = 0

    # 各要素が数値かどうかをチェック
    for item in lst:
        if isinstance(item, (int, float)):
            total += item
        else:
            is_numeric = False
            break

    # 数値のみのリストで、かつ合計が10以上の場合
    if is_numeric and total >= 10:
        result.append(lst)

print(result)
````

````python
# 発展形
# 数値のみのサブリストを抽出し、合計が10以上のものだけを新しいリストに格納
result = [lst for lst in data if all(isinstance(x, (int, float)) for x in lst) and sum(lst) >= 10]

print(result)
````
